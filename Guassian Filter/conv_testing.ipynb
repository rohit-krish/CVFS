```python
from convolution import convolution
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
```

```python
Kx = np.array(
    [[-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]], np.float32
)

Ky = np.array(
    [[1, 2, 1],
    [0, 0, 0],
    [-1, -2, -1]], np.float32
)

img = np.array(Image.open('../../assets/lenna.png'))

res = convolution(img, Kx)
```

```python
img = cv2.imread('../../assets/lenna.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
x_gradient = convolution(img,Kx)
y_gradient = convolution(img,Ky)
```

```python
plt.subplot(121)
plt.imshow(x_gradient,cmap='gray')
plt.subplot(122)
plt.imshow(y_gradient,cmap='gray')
plt.tight_layout()
plt.show()
```

```python
plt.imshow(np.hypot(x_gradient,y_gradient),cmap='gray')
```
