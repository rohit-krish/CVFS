```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2

plt.gray()
```

```python
img = Image.open('../assets/lenna.png')
img = np.array(img)
img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img = img[:,:-1]
if img.shape[1] % 2 == 1:
    img = img[:,:-1]

if img.shape[0] % 2 == 1:
    img = img[:-1,:]

plt.imshow(img)
plt.axis('off')
```

```python
_ = np.zeros(img.shape)
m,n = img.shape
res = _.copy()
# for i in range(img.shape[0]//2):
res[:,n//2:] = img[:,:n//2]
res[:,:n//2] = img[:,n//2:]
plt.axis('off')
plt.imshow(res)
```

```python
res = _.copy()
for i in range(n):
    res[:,i] = img[:,n-i-1]

plt.axis('off')
plt.imshow(res)
```

```python
res = _.copy()
for i in range(m):
    res[i,:] = img[m-i-1,:]

plt.axis('off')
plt.imshow(res)
```
